```python
# ANS 1.

import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

print("(i) Sum of all elements:", np.sum(gfg))
print("(ii) Sum of all elements row-wise:", np.sum(gfg, axis=1))
print("(iii) Sum of all elements column-wise:", np.sum(gfg, axis=0))
```

```python
# ANS 2.

import numpy as np

array = np.array([10, 52, 62, 16, 16, 54, 453])
print(f"(a) For array {array} :\n")
print("  (i) Sorted array:", np.sort(array))
print("  (ii) Indices of sorted array:", np.argsort(array))
print("  (iii) 4 smallest elements:", np.sort(array)[:4])
print("  (iv) 5 largest elements:", np.sort(array)[-5:])
print();

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print(f"(b) For array {array} :\n")
print("  (i) Integer elements only:",array[array == np.floor(array)])
print("  (ii) Float elements only:", array[array != np.floor(array)])

```

```python
# ANS 3.

import numpy as np


X = 148  # ASCII sum of A and S
sales = np.array([X, X+50, X+100, X+150, X+200])
print("(a) Sales:", sales)
print()


print("(b) Computing personalized tax rate ((X % 5) + 5) / 100 :")
tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * tax_rate
print("  (i) Tax rate:", tax_rate)
print("  (ii) Sales after tax:", taxed_sales)
print()


discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("(c) Discounted Sales:", discounted_sales)
print()


weeks = 3
weekly_sales = np.vstack([sales] * weeks)


week_multipliers = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
weekly_sales = weekly_sales * week_multipliers
print("(d) Weekly Sales after Increase:\n", weekly_sales)
```

```python
# ANS 4.

import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 100)


y1 = x**2                      
y2 = np.sin(x)                 
y3 = np.exp(x)                 
y4 = np.log(np.abs(x) + 1)     


plt.figure(figsize=(12, 8))

# GRAPH OF Y = x^2
plt.subplot(2, 2, 1)
plt.plot(x, y1, label=r"$Y = x^2$", color='r')
plt.title("Quadratic Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

# GRAPH OF Y = sin(x)
plt.subplot(2, 2, 2)
plt.plot(x, y2, label=r"$Y = \sin(x)$", color='g')
plt.title("Sine Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

# GRAPH OF Y = e^x
plt.subplot(2, 2, 3)
plt.plot(x, y3, label=r"$Y = e^x$", color='b')
plt.title("Exponential Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

# GRAPH OF Y = log(|x| + 1)
plt.subplot(2, 2, 4)
plt.plot(x, y4, label=r"$Y = \log(|x| + 1)$", color='purple')
plt.title("Logarithmic Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
